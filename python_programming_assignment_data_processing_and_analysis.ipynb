```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

hscore = max(employees.values())
print("Highest Score is: ",hscore)
topperformers = [name for name, score in employees.items() if score == hscore]
print(f"Top Performers Eligible for Bonus: {', '.join(topperformers)} (Score: {hscore})")
```

```python
import string
query = "Buy mobile phone buy phone online"
query = query.lower()
query = query.translate(str.maketrans('', '', string.punctuation))
words = query.split()
wordcount = {}
for word in words:
    wordcount[word] = wordcount.get(word, 0) + 1
repeated_keywords = {word: count for word, count in wordcount.items() if count > 1}
print(repeated_keywords)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_counts = {}
total_emails = len(emails)
for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```
